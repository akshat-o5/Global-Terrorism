## Years with most Terrorist Activity 

```python
df['Year'].value_counts().head(20)
```

```python
sns.set(rc={'figure.figsize':(15, 5)})
sns.barplot(y=df['Year'].value_counts()[:20].values, x=df['Year'].value_counts()[:20].index)
```

### From the above graph, we conclude that the most number of terrorist attack occured in 2014.

## Countries With Most Terrorist attacks 

```python
df['Country'].value_counts().head(10)
```

```python
sns.set(rc={'figure.figsize':(15, 5)})
sns.barplot(y=df['Country'].value_counts()[:10].values, x=df['Country'].value_counts()[:10].index)
```

### From the above graph, we conclude that the topmost country where most number of terrorist attack occured is Iraq, which is a hotspot/hotzone.
